```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
url = 'https://raw.githubusercontent.com/onaio/ona-tech/master/data/water_points.json'
data = pd.read_json(url)
df = pd.DataFrame(data)
df
```

```
#number of functional waterpoints
df = data.loc[:, 'water_point_condition':'communities_villages']
df
functional_waterpoints = df.groupby('water_point_condition')
functional_waterpoints.count()
```

from the output above, the number of functional waterpoints is 590

```
#number of waterpoints per community
water_point = df.groupby('communities_villages').water_point_condition.count()
water_point
```

```
#rank of each community by % of broken waterpoints
#no_of_broken_waterpoints = 40   #from the data above
df.sort_values("water_point_condition", inplace = True) 
filter = df["water_point_condition"]=="broken"
df.where(filter, inplace = True) 
df.count
```
